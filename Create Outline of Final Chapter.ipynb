```python
def aiwrite(n, temp, length):
  preamble = "### Context - This comes BEFORE your output paragraph:  " +summary + "\n ".join(short_prompts[0:n-1])
  prompt = "###  Take this sentence and expand it into a paragraph:  "+ short_prompts[n]
  postamble = " ".join(short_prompts[n+1:])
  postamble =  "### Context - This comes AFTER your output paragraph:  " + postamble


  messages = [
    {"role": "system", "content": "Answer consistent with the context."},
    {"role": "user", "content": prompt},
    {"role": "user", "content": preamble},
    {"role": "user", "content": postamble},
    {"role": "user", "content": "\n ### OUTPUT: \n"}
  ]

  response = openai.ChatCompletion.create(
    model="gpt-4",
    messages=messages,
    temperature=temp,
    max_tokens=length
  )

  content = response['choices'][0]['message']['content']
  return content

```

```python
def bulletize(content,n):
  sentences = sent_tokenize(content)
  example = "\n - ".join(sentences[n:])
  example = " - " + example
  return example

```
