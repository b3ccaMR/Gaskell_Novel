```python
chap_sum_1 = "In this chapter, the reader is introduced to Molly Gibson, a young girl who is excited to attend a festival at the Towers. The Towers is the home of Lord and Lady Cumnor, who are respected and influential figures in the town of Hollingford. Molly's father, Mr. Gibson, arranges for her to attend the festival, and she will be accompanied by the Miss Brownings. The chapter also mentions the presence of a widow named Clare, who is staying at the Towers and is in need of assistance."
```

```python
summaries = []
summaries.append(chap_sum_1)

i = 2
path = '/content/drive/My Drive/WiBD/Project_2/less data/chapters/'
for chapter in chapters:
    chap_name = path+ "wdchap"+str(i)+".txt"
    loader_short = TextLoader(chap_name)
    doc_short = loader_short.load()

    llm = ChatOpenAI(temperature=0, model_name="gpt-3.5-turbo-16k")
    chain = load_summarize_chain(llm, chain_type="stuff")
    summary = chain.run(doc_short)
    summaries.append(summary)
    print(summary)
    i += 1
```

```python
summary = "\n\n".join(summaries)
print(summary)

```

```python
filename = '/content/drive/My Drive/WiBD/Project_2/less data/langchain_chapter_summaries.txt'
loader = TextLoader(filename)
documents = loader.load()
text_splitter = CharacterTextSplitter(chunk_size=1000, chunk_overlap=0)
texts = text_splitter.split_documents(documents)

embeddings = OpenAIEmbeddings()
docsearch = Chroma.from_documents(texts, embeddings)

qa = RetrievalQA.from_chain_type(llm=OpenAI(), chain_type="stuff", retriever=docsearch.as_retriever())
```

```python
query = "Who is Molly's step sister?"
qa.run(query)
```
